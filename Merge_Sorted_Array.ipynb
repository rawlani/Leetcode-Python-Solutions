**Merge Sorted Array**<br>
Given two sorted integer arrays nums1 and nums2, merge nums2 into nums1 as one sorted array.<br>

**Note:**<br>
The number of elements initialized in nums1 and nums2 are m and n respectively.<br>
You may assume that nums1 has enough space (size that is equal to m + n) to hold additional elements from nums2.

**Example:**

**Input:**<br>
nums1 = [1,2,3,0,0,0],<br>
m = 3<br>
nums2 = [2,5,6],<br>
n = 3<br>

**Output:** [1,2,2,3,5,6]

**Concept 1:** Going from Left to Right

For this:
1. Make a copy of nums1 list before 0s
2. Empty all the elements of nums1
3. Create two pointers (one for nums1_copy and other of nums2) with value 0 to keep a track of elements added in nums1
4. Iterate until the one or both pointer becomes more than the count of m and n

**Example:**<br>
(Initial starting point)<br>
nums1_copy = [1,2,3]<br>
nums2 = [2,5,6]<br>
nums1 = [ ]<br>
p1 = 0 , m = 3<br>
p2 = 0 , n = 3<br>

    Iteration 1:
        1 < 2, so 1 gets added in nums1 list, p1 becomes 1

    Iteration 2:
        2 < 2, else condition -> 2 gets added in nums2, p2 becomes 1
    
    Iteration 3:
        2 < 5, so 2 gets added in nums1 list, p1 becomes 2

    Iteration 4:
        3 < 5, so 3 gets added in nums1 list, p1 becomes 3

Iteration ends because p1 is now not less than m

So, remaining numbers are directly just appended as it is in same order<br>
In this case p2 < n, so **nums1[p1+p2:] = nums2[p2:]**

```python
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        
        nums1_copy = nums1[:m]
        nums1[:] = []
        p1 = 0
        p2 = 0
        
        while p1 < m and p2 < n:
            if(nums1_copy[p1] < nums2[p2]):
                nums1.append(nums1_copy[p1])
                p1 = p1 + 1
            else:
                nums1.append(nums2[p2])
                p2 = p2 + 1
                
        if(p2 < n):
             nums1[p1+p2:] = nums2[p2:]
        if(p1 < m):
            nums1[p1+p2:] = nums1_copy[p1:]
```

**Concept 2:** Going from Right to Left<br>
In this case, we need not to create any extra space and we will have an in-place manipulation

For this:
1. Create three pointers where<br>
    p1 = len(nums1)-1   or   p1 = m+n-1<br>
    p2 = len(nums2)-1   or   p2 = n-1<br>
    p  = m - 1<br><br>
2. Iterate until p2 and p are greater than 0

**Remember:**<br>
(Initial pointers)<br>
1. p1 is pointing at nums1's last element<br>
2. p2 is pointing at nums2's last element<br>
3. p is pointing at nums1's last non-zero element

**Example:**<br>
(Initial starting point)<br>
nums1 = [1,2,3,0,0,0]<br>
nums2 = [2,5,6]<br>
p1 = 5 , m = 3<br>
p2 = 2 , n = 3<br>
p = 2

    Iteration 1:
        3 < 6, so 6 gets in-placed with nums1[5], p2 becomes 1, p1 becomes 4
    
    Iteration 2:
        3 < 5, so 5 gets in-placed with nums1[4], p2 becomes 0, p1 becomes 3
       
    Iteration 3:
        3 < 2, else condition -> 3 gets in-placed with nums1[3], p becomes 1, p1 becomes 2
        
    Iteration 4:
        2 < 2, else condition -> 2 gets in-placed with nums1[2], p becomes 0, p1 becomes 1
        
Remaining directly just gets appended up

```python
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        
        p1 = m+n-1
        p2 = n-1
        p = m-1
        
        while p2 >= 0 and p >= 0:
            if(nums1[p] < nums2[p2]):
                nums1[p1] = nums2[p2]
                p2 = p2 - 1
                p1 = p1 - 1
            else:
                nums1[p1] = nums1[p]
                p = p - 1
                p1 = p1 - 1
        while p2>=0:
            nums1[p1] = nums2[p2]
            p1 = p1 - 1
            p2 = p2 - 1
```
