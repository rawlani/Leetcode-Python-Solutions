**Given a roman numeral, convert it to an integer.**

**Example 1:**

Input: s = "III"<br>
Output: 3

**Example 2:**

Input: s = "LVIII"<br>
Output: 58<br>
Explanation: L = 50, V= 5, III = 3.

**Concept 1:** Reading from Left to Right and adding accordingly

Explaination:

    Example: 'MCMXCIV'
    Take 'value' as a variable which store total sum
        1st Iteration:
            M = 1000
            So, value = 1000
        
        2nd Iteration:
            C = 100
            value = 1000 + 100
         
        3rd Iteration:
            (check the prev value was smaller than present one)
            M = 1000
            value = 1000 + 100 + 1000 - 2(100) = 1900
             
         4th Iteration:
            X = 10
            value = 1900 + 10
           
         5th Iteration:
             (check the prev value was smaller than present one)
             C = 100
             value = 1900 + 10 + 100 - 2(10) = 1990
          
         6th Iteration:
             I = 1
             value = 1990 + 1
            
         7th Iteration:
             (check the prev value was smaller than present one)
             V = 5
             value = 1990 + 1 + 5 - 2(1) = 1994
         
*This way Roman is converted to Integer*

**Note:**
    Whenever the current value was larger than the previous one, then twice the smaller value was subtracted from the total value.
              


```python
# Runtime: 40 ms
# Memory Usage: 14 MB

class Solution:
    def romanToInt(self, s: str) -> int:
        
        r2i_dict = {'I':1, 'V':5, 'X':10, 'L':50, 'C':100, 'D':500, 'M':1000}
        r2i = 0
        for i in range(len(s)):
            try:
                if(r2i_dict[s[i]] < r2i_dict[s[i+1]]):
                    r2i = r2i - r2i_dict[s[i]]
                else:
                    r2i = r2i + r2i_dict[s[i]]
            except:
                r2i = r2i + r2i_dict[s[i]]
        return r2i
```

**Concept 2:** Reading from Right to Left and adding accordingly

Explaination:

    Example: 'LIX'
    Take 'value' as a variable which store total sum
    
    Note: Before starting we need to reverse it so that we can start reading from backwards
              So, now after reversing the example becomes 'XIL'
    
        1st Iteration:
            X = 10
            So, value = 10

        2nd Iteration:
            (check the prev value is bigger than present one)
            I = 1
            value = 10 - 1 = 9

        3rd Iteration:
            (check the prev value was bigger than present one)
            L = 50
            value = 9 + 50 = 59
            
*This way Roman is converted to Integer*

```python
# Runtime: 32 ms
# Memory Usage: 14 MB
    

class Solution:
    def romanToInt(self, s: str) -> int:
        
        r2i_dict = {'I':1, 'V':5, 'X':10, 'L':50, 'C':100, 'D':500, 'M':1000}
        r2i = 0
        prev = 0
        s = s[::-1]
        for i in range(len(s)):
            if(r2i_dict[s[i]] < prev):
                r2i = r2i - r2i_dict[s[i]]
            else:
                r2i = r2i + r2i_dict[s[i]]
            prev = r2i_dict[s[i]]
    return r2i
```
